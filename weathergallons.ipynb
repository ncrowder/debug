```python
import pandas as pd
import pyodbc 

conn = pyodbc.connect('Driver={SQL Server};'
                      'Server=KTDBBIPRD;'
                      'Database=BIDEV;'
                      'Trusted_Connection=yes;')

weatherdf = pd.read_sql_query('SELECT * FROM stg.WeatherCurrentAndHistorical WHERE convert(date,CurrentTime) between \'2020-12-01\' and \'2021-04-01\'', conn)
```

```python
weatherdf.head(10)
```

```python
weatherdf.columns
```

```python
weatherdf = weatherdf.filter(['Id','SiteNumber','ZipCode','City','State','Latitude','Longitude','CurrentTime','Temperature','FeelsLike','Rain','Snow'])
```

```python
weatherdf.head()
```

```python
groupbydf = weatherdf.groupby(['SiteNumber',weatherdf.CurrentTime.dt.date]).agg({"Temperature":'min',"Snow":'sum'})
```

```python
#convert mm to inches
groupbydf.Snow = groupbydf.Snow/25.4
```
